```python
import psycopg2
```

```python
## Database connection parameters
conn = psycopg2.connect(
    dbname="bda_db",
    user="postgres",
    password="pwd",
    host="localhost",  # Change to your server if needed 127.0.0.1
    port="5433"  # Default PostgreSQL port
)
```

```python
## Create a cursor to execute queries
cur = conn.cursor()
```

```python
## Execute a simple query
cur.execute("SELECT * from users_tbl")
```

```python
print(cur.fetchall())
```

```python
## Fetch and print the result
## This method retrieves the next row of a query result set and returns a single sequence, or None if no more rows are available.
print(cur.fetchone())
```

```python
print(cur.fetchone())
```

```python
print(cur.fetchone())
```

```python
print(cur.fetchone())
```

```python
## Close connection
cur.close()
conn.close()
```
